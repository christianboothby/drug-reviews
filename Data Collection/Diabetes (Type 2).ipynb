```python
import requests
import bs4
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from tqdm import tqdm
from time import time, sleep
```

```python
url_template = "https://www.drugs.com/comments/dulaglutide/trulicity-for-diabetes-mellitus-type-ii.html?page={}"
max_results_per_page = 32
rating = []
review = []
date = []
taken = []
likes = []

for start in tqdm(range(max_results_per_page)):
        r1 = requests.get(url_template.format(start))
        soup = BeautifulSoup(r1.text, 'html.parser')
        start = time()
        while time()-start<1: continue
     
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                rating.append(x.find('div', attrs={'class': 'ddc-rating-summary'}).find('b').text)
            except:
                rating.append(np.nan)
            
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                review.append(x.find('p', attrs={'class': 'ddc-comment-content'}).text.strip())
            except:
                review.append(np.nan)
            
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                date.append(x.find('div', attrs={'class': 'ddc-comment-section'}).find('span', attrs={'class', 'comment-date'}).text)
            except:
                date.append(np.nan)
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                likes.append(x.find('div', attrs={'class': 'ddc-comment-actions'}).next_sibling.next_sibling.find('a').text)
            except:
                likes.append(np.nan)
            
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                taken.append((x.find('div', attrs={'class': 'ddc-comment-section'}).find('span', attrs={'class': 'ddc-text-color-secondary'}).next_sibling.next_sibling).text)
            except:
                taken.append(np.nan) 
```

```python
df = pd.DataFrame({'Rating': rating,
            'Date': date,
            'Taken': taken,
            'Helpful': likes,
            'Review': review})

df['Drug'] = 'Trulicity'
df['Condition'] = 'Diabetes (Type 2)'
df['Major Interaction'] = 2
df['Moderate Interaction'] = 216
df['Minor Interaction'] = 22
df['Dosage'] = 'injection'
df.drop_duplicates(inplace=True)
df
```

```python
df10 = pd.DataFrame({'Rating': rating,
            'Date': date,
            'Taken': taken,
            'Helpful': likes,
            'Review': review})

df10['Drug'] = 'Glimepiride'
df10['Condition'] = 'Diabetes (Type 2)'
df10['Major Interaction'] = 19
df10['Moderate Interaction'] = 388
df10['Minor Interaction'] = 34
df10['Dosage'] = 'oral'
df10.drop_duplicates(inplace=True)
df10
```
