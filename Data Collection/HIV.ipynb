```python
import requests
import bs4
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from tqdm import tqdm
from time import time, sleep
```

```python
url_template = "https://www.drugs.com/comments/cobicistat-elvitegravir-emtricitabine-tenofovir-alafenamide/genvoya-for-hiv-infection.html?page={}"
max_results_per_page = 6
rating = []
review = []
date = []
taken = []
likes = []

for start in tqdm(range(max_results_per_page)):
        r1 = requests.get(url_template.format(start))
        soup = BeautifulSoup(r1.text, 'html.parser')
        start = time()
        while time()-start<1: continue
     
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                rating.append(x.find('div', attrs={'class': 'ddc-rating-summary'}).find('b').text)
            except:
                rating.append(np.nan)
            
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                review.append(x.find('p', attrs={'class': 'ddc-comment-content'}).text.strip())
            except:
                review.append(np.nan)
            
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                date.append(x.find('div', attrs={'class': 'ddc-comment-section'}).find('span', attrs={'class', 'comment-date'}).text)
            except:
                date.append(np.nan)
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                likes.append(x.find('div', attrs={'class': 'ddc-comment-actions'}).next_sibling.next_sibling.find('a').text)
            except:
                likes.append(np.nan)
            
    
        for x in soup.find_all('div', attrs={'class': 'ddc-comment'}):
            try:
                taken.append((x.find('div', attrs={'class': 'ddc-comment-section'}).find('span', attrs={'class': 'ddc-text-color-secondary'}).next_sibling.next_sibling).text)
            except:
                taken.append(np.nan) 
```

```python
df = pd.DataFrame({'Rating': rating,
            'Date': date,
            'Taken': taken,
            'Helpful': likes,
            'Review': review})

df['Drug'] = 'Genvoya'
df['Condition'] = 'HIV/AIDS'
df['Major Interaction'] = 285
df['Moderate Interaction'] = 258
df['Minor Interaction'] = 18
df['Dosage'] = 'oral'
df.drop_duplicates(inplace=True)
df
```

```python
df10 = pd.DataFrame({'Rating': rating,
            'Date': date,
            'Taken': taken,
            'Helpful': likes,
            'Review': review})

df10['Drug'] = 'Efavirenz'
df10['Condition'] = 'HIV/AIDS'
df10['Major Interaction'] = 219
df10['Moderate Interaction'] = 442
df10['Minor Interaction'] = 13
df10['Dosage'] = 'oral'
df10.drop_duplicates(inplace=True)
df10.head()
```
